```python
produtos = {
    'Produto': [
        'Notebook',
        'Mouse',
        'Teclado',
        'Monitor',
        'Headset',
        'Impressora',
        'Webcam',
        'SSD'
    ],

    'Categoria': [
        'Informática',
        'Periféricos',
        'Periféricos',
        'Informática',
        'Periféricos',
        'Informática',
        'Periféricos',
        'Armazenamento'
    ],

    'Valor': [
        3500,
        80,
        150,
        1200,
        250,
        900,
        180,
        400
    ],

    'Estoque': [
        15,
        120,
        80,
        25,
        40,
        10,
        50,
        35
    ]
}

df = pd.DataFrame(produtos)
df_preco = df.sort_values(by='Valor')
print(df_preco)
df_preco_decrescente = df.sort_values(by='Valor',ascending=False)
print(df_preco_decrescente)

df['Valor total'] = df['Valor'] * df['Estoque']
print(df)
df_novo = df.rename(columns={'Valor' : 'Preço'})
print(df_novo)
```

```python
# ==========================================
# EXERCÍCIO 1 - SERIES
# ==========================================

# Crie uma Series contendo:

# 10
# 20
# 30
# 40
# 50

# Mostre:
# - A Series completa
# - O primeiro elemento
# - O último elemento

Numeros = [10,20,30,40,50]
series = pd.Series(Numeros)
print("Exercício 1: ")
print("Series completa: ")
print(series)
print("Primeiro elemento: ",series[0])
print("último elemento: ",series.iloc[-1])

# ==========================================
# EXERCÍCIO 2 - SERIES COM ÍNDICES
# ==========================================

# Crie uma Series:

# a -> 100
# b -> 200
# c -> 300

# Mostre:
# - A Series completa
# - O valor associado ao índice "b"

numeros = [100,200,300]
indices = ['a','b','c']

print("Exercício 2: ")
series = pd.Series(numeros,index=indices)
print(series)
print("Número associado ao índice 'b: " ,series['b'])
# ==========================================
# EXERCÍCIO 3 - SERIES A PARTIR DE DICIONÁRIO
# ==========================================

# Crie uma Series utilizando:

# nome -> Ana
# idade -> 25
# cidade -> São Paulo

# Mostre a Series.

print("Exercício 3: ")
dicionario = {
    'nome' : 'Ana',
    'idade' : 25,
    'cidade' : 'São Paulo'
}
infos = pd.Series(dicionario)
print(infos)

# ==========================================
# EXERCÍCIO 4 - DATAFRAME
# ==========================================

# Crie um DataFrame com:

# Nome
# Ana
# Bruno
# Carlos

# Idade
# 25
# 30
# 19

# Cidade
# São Paulo
# Campinas
# Curitiba

# Mostre o DataFrame.
informações = {
    'Nome' : ['Ana','Bruno','Carlos'],
    'Idade': [25, 30, 19],
    'Cidade':['São Paulo', 'Campinas', 'Curitiba']
}

print("Exercício 4: ")
dataframe = pd.DataFrame(informações)
print(dataframe)
# ==========================================
# EXERCÍCIO 5 - ACESSANDO COLUNAS
# ==========================================

# Utilizando o DataFrame anterior:

# Mostre:
# - Apenas a coluna Nome
# - Apenas a coluna Cidade
# - As colunas Nome e Idade

print("Exercício 5: ")
print("Nomes: ")
print(dataframe['Nome'])
print("Cidades: ",dataframe['Cidade'])
print("Nome e Idade: ",dataframe[['Nome','Idade']])


# ==========================================
# EXERCÍCIO 6 - LOC
# ==========================================

# Crie:

# Produto
# Notebook
# Mouse
# Teclado

# Preco
# 3500
# 80
# 150

# Estoque
# 10
# 50
# 30

# Mostre:
# - A primeira linha usando loc
# - Apenas Preco e Estoque da segunda linha

Produtos = {
    'Produto' : ['Notebook','Mouse','Teclado'],
    'Preço' : [3500,80,150],
    'Estoque' : [10,50,30],
}

print("Exercício 6: ")
infoprod = pd.DataFrame(Produtos)
print("Primeira linha: ",infoprod.loc[0])
print("Preço e Estoque da segunda linha: ",infoprod.loc[1,['Preço', 'Estoque']])

# ==========================================
# EXERCÍCIO 7 - FILTRO SIMPLES
# ==========================================

# Utilizando o DataFrame de produtos:

# Mostre apenas produtos:
# - Com preço maior que 100
# - Com estoque maior que 20


precos_maiores = infoprod['Preço'] > 100
estoque_maior = infoprod['Estoque'] > 20
print("Exercício 7: ")
print("Produtos com preço maior que 100: ",infoprod[precos_maiores])
print("Produtos com estoque maior que 20: ",infoprod[estoque_maior])

# ==========================================
# EXERCÍCIO 8 - FILTRO COMPOSTO
# ==========================================

# Mostre apenas produtos:

# Preco > 100
# E
# Estoque > 20

print("Exercício 8: ")
print("Produtos com preço maior que 100 e estoque maior que 20: ")
print(infoprod[(infoprod['Preço'] > 100) & (infoprod['Estoque'] > 20)])

# ==========================================
# EXERCÍCIO 9 - FILTRO COM OR
# ==========================================

# Mostre produtos:

# Estoque < 20
# OU
# Preco > 1000

print("Exercício 9: ")
print("Produtos com preço maior que 1000 ou estoque menor que 20: ")
print(infoprod[(infoprod['Preço'] > 1000) | (infoprod['Estoque'] < 20)])


# ==========================================
# EXERCÍCIO 10 - SORT VALUES
# ==========================================

# Ordene o DataFrame:

# - Do produto mais barato para o mais caro
# - Do produto mais caro para o mais barato

print("Exercício 10: ")
print("Produto do mais barato para o mais caro: ")
print(infoprod.sort_values(by='Preço'))
print("Produto do mais caro para o mais barato: ")
print(infoprod.sort_values(by='Preço',ascending=False))

# ==========================================
# EXERCÍCIO 11 - ASTYPE
# ==========================================

# Crie um DataFrame:

# Nome      Idade
# Ana       "25"
# Bruno     "30"
# Carlos    "19"

# Converta a coluna Idade para inteiro.

dadospessoas = {
    'Nome' : ['Ana','Bruno','Carlos'],
    'Idade' : ['25','30','19']
}

df = pd.DataFrame(dadospessoas)
print("Exercíci0 11: ")
df['Idade'] = df['Idade'].astype(int)
print(df)
print(df['Idade'].dtype)

# ==========================================
# EXERCÍCIO 12 - ISNA
# ==========================================

# Crie um DataFrame:

# Nome      Idade
# Ana       25
# Bruno     NaN
# Carlos    19

# Mostre quais valores são nulos.

pessoas = {
    'Nome' : ['Ana','Bruno','Carlos'],
    'Idade' : [25, np.nan ,19]
}

print("Exercício 12: ")
df2 = pd.DataFrame(pessoas)
print(df2.isna())


# ==========================================
# EXERCÍCIO 13 - MERGE
# ==========================================

# Crie:

# df_clientes
#
# ID   Nome
# 1    Ana
# 2    Bruno

# df_compras
#
# ID   Produto
# 1    Notebook
# 2    Mouse

# Faça um merge pela coluna ID.

print("Exercício 13: ")
df_clientes = pd.DataFrame({
    'ID': [1, 2],
    'Nome': ['Ana', 'Bruno']
})

df_compras = pd.DataFrame({
    'ID': [1, 2],
    'Produto': ['Notebook', 'Mouse']
})

df_final = pd.merge( df_clientes, df_compras, on='ID')

print(df_final)
```
