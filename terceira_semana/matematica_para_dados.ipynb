```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.stats.proportion import proportions_ztest
```

**Estatística**



*   População: conjunto completo de elementos que estamos analisando.
*   Amostra: parte representativa da população. Quando não dá para analisar todo mundo.
*   Variáveis: características que você mede ou observa nos indivíduos. As variaveis podem ser qualitativas (representam categorias ou características) ou quantitativas(representam valores numéricos).
*   Medidas de tendência central: medidas que resumem um conjunto de dados. Podendo ser média, moda ou mediana.
*   Média aritmética: soma dos valores / quantidade de valores.
*  Moda: valor que mais se repete em um conjunto de dados.
*   Mediana: valor central de um conjunto de dados ordenado.
*  Amplitude: diferença entre o maior e o menor valor do conjunto de dados. Menor amplitude = dados mais concentrados.
*   Desvio padrão: mostra o quanto, em média, os valores se afastam do centro, da média. Raiz quadrada da variância.










**Medidas de posição**



*   Quartis: Q1 = 25%, Q2= 50%/mediana, Q3 = 75%, Q4 = 100%.
*   fórmula do quartil para dados agrupados em classes (com intervalo):
Qk​= Limite inferior da classe + (Posição do quartil - Frequência acumulada inferior / frequencia da classe) x amplitude da classe.
*   Amplitude interquantil: Q3 - Q1
*   Boxplot: Precisa ter Q1, mediana/Q2, Q3, amplitude interquartil, além disso é necessário calcular o limite inferior e o superior.
Limite superior: Q3 + 1,5 x IQR.
Limite inferior: Q1 - 1,5 x IQR.
*   Valores menores ou maiores que os limites são chamados de outliers.

**Análise combinatória e probabilidade**

*   PFC: "e" = p1 x p2 x p3...pn; "ou" = p1 + p2 + pn
*   Fatorial: n!=nx(n-1)x(n-2)x⋯x1
*   Permutação simples: mudar a ordem de n objetos distintos. n!
*   Arranjo simples: ordem importa e apenas alguns elementos são escolhidos. Fórmula = n!/(n - p)!
*   Combinação simples: escolher p entre n objetos. Ordem não importa. Fórmula = n!/p!(n - p)!
*   Espaço amostral: conjunto de todos os resultados possíveis.
*   Evento: subconjunto do espaço amostral. ex: ocorrer face par.
*   Cálculo de probabilidade = Eventos / Espaço amostral.
*   Probabilidade condicional: probabilidade de A acontecer, dado que B já ocorreu.
*   Distribuição normal: se os dados parecem seguir a distribuição normal, então, é simétrico, com a maioria dos dados seguindo a média e poucos valores extremos. Assim, média = mediana = moda.
*   Intervalo de confiança: faixa de valores que provavelmente contém o valor verdadeiro da população.
Fórmula = média da amostra +/- (valor crítico x desvio padrão)/raiz do tamanho da amostra.
*   Teste de hipótese: confirmar se determinado dado é confiável. Há a hipótese nula que assumimos que é verdadeira e outra, que é alternativa. Também temos um valor de tolerância para erros.


**Álgebra Linear**

*  Matrizes: tabela de números, organizada em linhas e colunas.



**Análise matemática usando um dataset**

```python
url="https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv?utm_source=chatgpt.com"
titanic = pd.read_csv(url)
titanic.head()
```

```python
media = titanic['Age'].mean()
mediana = titanic['Age'].median()
moda = titanic['Age'].mode()[0]
minimo = titanic['Age'].min()
maximo = titanic['Age'].max()
amplitude = maximo - minimo
variancia = titanic['Age'].var()
desvio_padrao = titanic['Age'].std()

print("Estatística descritivas e de dispersão:")
print(f"A média de idade dos passageiros é: {media:.2f} anos")
print(f"A mediana de idade dos passageiros é: {mediana:.2f} anos")
print(f"A moda da idade dos passageiros é: {moda:.2f} anos")
print(f"O minimo das idades é: {minimo} e o maximo é: {maximo}")
print(f"A amplitude da idade dos passageiros é: {amplitude} anos")
print(f"A variância da idade dos passageiros é: {variancia:.2f}")
print(f"O desvio padrão da idade dos passageiros é: {desvio_padrao:.2f}")
q1 = titanic['Age'].quantile(0.25)
q2 = titanic['Age'].quantile(0.50)
q3 = titanic['Age'].quantile(0.75)
iqr = q3 - q1
lim_inf = q1 - 1.5 * iqr
lim_sup = q3 + 1.5 * iqr
print("1° quartil: ",q1)
print("2° quartil: ",q2)
print("3° quartil: ",q3)
print("IQR = ", iqr)
print("Limite inferior: ", lim_inf)
print("Limite superior: ", lim_sup)
descricao = titanic["Age"].describe()
print(descricao)
```

**Gráficos**

```python
# Histograma das idades dos passageiros:
plt.hist(titanic['Age'].dropna(), bins=25)
plt.title('Variação das idades dos passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
# O gráfico indica que a distribuição de idades ficou concentrada na faixa
# entre 20 e 30 anos, mas existem passageiros distribuidos por todas as faixas etarias.
```

```python
# Boxplot das idades dos passageiros
plt.boxplot(titanic['Age'].dropna())
plt.title('Boxplot das Idades dos passageiros')
plt.show()
# O boxplot mostra algumas medidas que foram estudadas anteriormente:
# o primeiro quartil (Q1) está próximo de 20 anos, o segundo quartil
# (Q2), é correspondente à mediana e está próximo de 28 anos, e o terceiro
# quartil (Q3) está próximo de 38 anos. Indicando que cerca de 50% dos
# passageiros possuíam idade entre os 20 anos a 38 anos.
# Observa-se também uma grande amplitude nas idades dos passageiros,
# que possuíam menos de 1 ano até os 80 anos.
# Também, há a presença de outliers, que são observações localizadas
# fora dos limites - inferiores ou superiores - definidos pelo boxplot.
# Nesse conjunto de dados, os outliers correspondem a passageiros com idades
# superiores a aproximadamente 65 anos.
```

**Intervalo de confiança**

```python
idades = titanic['Age'].dropna() # Remove os valores nulos
media = idades.mean() # Média
desvio = idades.std() # Desvio padrão
n = len(idades) # Tamanho da amostra
z = 1.96 # Valor crítico para intervalo de confiança de 95%
margem_erro = z * (desvio / np.sqrt(n)) # Margem de erro
limite_inferior = media - margem_erro
limite_superior = media + margem_erro
print(f"Média: {media:.2f}") # Média das idades
print(f"Desvio padrão: {desvio:.2f}") # Dispersão considerável das idades
print(f"Tamanho da amostra: {n}")
print(f"Margem de erro: {margem_erro:.2f}") # Incerteza da estimativa.
print(f"Intervalo de confiança de 95%: [{limite_inferior:.2f}, {limite_superior:.2f}]") # Intervalo de confiança = média +/- margem de erro
```

**Teste de hipótese**

```python
# Hipótese nula: a maioria dos passageiros do titanic morreu? (P = 0.5)
# Hipótese alternativa: menos de 50% sobreviveram (P < 0.5)
# Sendo P a probabilidade de ter sobrevivido.
sobreviventes = titanic['Survived'].sum()
passageiros = len(titanic)

estatistica, pvalor = proportions_ztest(
    count=sobreviventes, # número de sobreviventes.
    nobs=passageiros, # total de pessoas no titanic.
    value=0.5, # hipótese nula: 50% das pessoas sobreviveram.
    alternative='smaller' # hipótese alternativa: menos de 50% sobreviveram.
)

print(f"Sobreviventes do Titanic: {sobreviventes}")
print(f"Total de passageiros no Titanic: {n}")
print(f"Taxa de sobrevivência: {sobreviventes/passageiros:.2%}")
print(f"Z: {estatistica:.4f}") # Quão longe o resultado observado está da hipótese nula.
print(f"p-valor: {pvalor:.8f}") # Qual a chance de observarmos um resultado igual a esse se a hipótese nula fosse verdadeira.
```

Conclusão: A análise desse dataset mostrou que a idade média dos passageiros era de aproximadamente 29,7 anos, com desvio padrão de 14,53 anos, ou seja, tem uma grande variação da idade dos passageiros. O intervalo de confiança de 95% indicou que a idade média populacional está entre 28,63 e 30,76 anos, bem próxima da idade média. O teste de hipótese foi realizado para avaliar se a taxa de sobrevivência dos passageiros do Titanic era igual a 50%(hipótese nula). Os resultados indicaram uma estatística Z de -7,13 e p-valor menor que 0,05. Assim, rejeitou-se a hipótese nula, concluindo-se que a taxa de sobrevivência foi muito inferior a 50%, mostrando que a maioria dos passageiros não sobreviveu ao acidente.
